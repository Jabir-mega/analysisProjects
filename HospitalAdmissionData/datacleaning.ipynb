```python
# we make sure to strip the heading so as to prevent any errors from mismatch btwn the two 
df2.columns = df2.columns.str.strip()
df.columns = df.columns.str.strip()
```

```python
# write a code to transfer full forms of headings from df2 to the abbreviations in df
# step 1 - put the file responsible for mapping into a dictionary form
# within the zip() we create two list and then combine them pair by pair 
df2_dictionary = dict(zip(df2["Table Heading"], df2["Explanatory Name"]))
```

```python
# tried using the conventional .strip() method but to no avail due to dict dtype
df2_dictionary = {k.strip(): v.strip() for k, v in df2_dictionary.items()}
```

```python
df.rename(columns=df2_dictionary, inplace=True)
```

```python
# Cleaning numerical columns 
# implement a function
def clean_num_col(df, columns):
    for col in columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df
```

```python
numerical_columns = []
def return_int(df):
    for col in df.columns:
        if df[col].dtypes == "int64":
            numerical_columns.append(col)
    return numerical_columns
            
return_int(df)
numerical_columns
```

```python
df1_cleaned = clean_num_col(df, numerical_columns)
```

```python
df1_cleaned
```

```python
# format date time column
def clean_date(df, columns):
    for col in columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
    return df

date_columns = ['Date of Admission', 'Date of Discharge']
df1_cleaned = clean_date(df, date_columns)
```
